# Inverse bilinear transformation in 2D

```python
import sympy as sy
sy.init_printing()

[x11, x12, x21, x22, y11, y12, y21, y22, u, v, xx, yy] = sy.symbols('x11 x12 x21 x22 y11 y12 y21 y22 u v xx yy')
x = (1-u) * (1-v) * x11 + u * (1-v) * x21 + (1-u) * v * x12 + u * v * x22
y = (1-u) * (1-v) * y11 + u * (1-v) * y21 + (1-u) * v * y12 + u * v * y22
x
```

```python
y
```

```python
tmp = sy.solve([x-xx,y-yy],[u,v])
tmp 
```

```python
sy.simplify(tmp[1])
# sol
```
